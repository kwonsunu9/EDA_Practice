#### 패키지 import

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

#### 데이터 로드

```python
survey = pd.read_csv('./kaggle_survey_2022_responses.csv')
survey
```

#### 컬럼 분류

```python
per_col = ["Q2", "Q3", "Q4", "Q5", "Q29", "Q30"]
ds_col = ["Q11", "Q16", "Q23"]
edu_col = ["Q8", "Q9"]
com_col = ["Q24", "Q25", "Q26", "Q27"]
```

# personally column

```python
edit_data = survey[per_col]
edit_data
```

```python
edit_data[edit_data.isnull().any(axis=1)]
```

```python
final_data = edit_data.dropna()
final_data
```

* Q2, Q3, Q4, Q5까지 기본적인 질문에 대한 작업

* Q30, Q40은 개인의 급여와 지출에 대한 범주형 자료 분석을 실행

### 연령대

#### 연령대 빈도 수

```python
Q2 = final_data["Q2"][1:]
Q2.value_counts()
```

* 20대 후반에서 40대까지 많은 빈도를 보이고 있음.
* 계급 별 빈도를 미루어 볼 때 왼쪽으로 치우친 가우시안 분포에 가까울 것으로 기대가 됨.

#### 연령대 시각화

```python
plt.figure(figsize=(16, 12))
ord = ['18-21', '22-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60-69', '70+']
sns.countplot(x = "Q2", data = final_data[1:], order = ord).set(title = 'age_range distribution')
```

```python
plt.figure(figsize=(16, 16))
explode = [0, 0.10, 0.2, 0.30, 0.4, 0.5, 0.6, 0.7,0.8, 0.9, 1]
plt.pie(Q2.value_counts(),
       labels=Q2.value_counts().index,
       autopct='%d%%',
       startangle=90,
       textprops={'fontsize':12}, explode=explode)
plt.axis('equal')
plt.title("Pie chart for Q2 column", fontsize=16)
```
